```python
class Listener(StreamListener):
    
    def on_data(self, data):
        tweet = json.loads(data)
        try:
            text = tweet['text'].lower()
            text = re.sub('\[.*?\]', '', text)
            text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
            text = re.sub('\w*\d\w*', '', text)
            text = re.sub('[‘’“”…]', '', text)
            text = re.sub('\n', '', text)

            text = " ".join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)"," ", text).split())
            text = " ".join(re.sub("rt"," ", text).split())

            global col1
            global col2
            cl1 = 0
            cl2 = 0
            blob = TextBlob(text.strip())
            for sentence in blob.sentences:
                #cl1 = sentence.sentiment.polarity
                #cl2 = sentence.sentiment.subjectivity
                if "trump" in sentence:
                    cl1 = sentence.sentiment.polarity
                else:
                    cl2 = sentence.sentiment.polarity
            
            col1 += cl1
            col2 += cl2
            #print(pol,sub)
            #df.loc[k] = [pol,sub]
            #k += 1
            with open('election.csv','a') as file:
                writer = csv.DictWriter(file, fieldnames=header_name)
                info = {'col1':col1,'col2':col2}
                writer.writerow(info)
        except:
            print('Error')
       
    
    def on_error(self, status):
        print(status)
```
